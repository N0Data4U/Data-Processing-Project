Building on the scraper we built for getting all the data for one product page we want now automize the process for all product pages.

```python
import requests
from bs4 import BeautifulSoup

# 1 Finding website with all the products listed
category_url = "https://www.datart.cz/mobilni-telefony.html?limit=1000" # setting limit manually in the browser, so every product appears
basic_url = "https://www.datart.cz"
# Defining a function 
def get_product_urls(category_url):
    response = requests.get(category_url)
    if response.status_code == 200: # if 200, then its possible to get data 
        soup = BeautifulSoup(response.text, 'html.parser')
        # Extract product URLs using appropriate selectors
        product_urls = [[basic_url + item.a.get('href')] for item in soup.select('.item-title')]
        
        return product_urls
    else:
        print("Failed fetching")
        return []

product_urls = get_product_urls(category_url)
```
